```python
#Compute differences map
import xarray as xr
import numpy as np
import os
import glob
```

```python
import matplotlib.pyplot as plt

# Quick plot
nc_files = sorted(glob.glob('/Users/epauthen/Documents/GitHub/Hackathon-Meteo-France/Isotherme0_data/freezing_days_per_year*.nc'))
ds = xr.open_mfdataset(nc_files, decode_times=False).load()
# Select data for the two periods
period_early = ds.freezing_days.sel(year=slice(1990, 2020))
period_late = ds.freezing_days.sel(year=slice(2080, 2100))

# Compute the mean over each period
mean_early = period_early.mean(dim="year")
mean_late = period_late.mean(dim="year")

# Compute the difference: late period minus early period
diff_map = mean_late - mean_early

# Optional: quick plot for visualization
plt.figure()
diff_map.plot(cmap='Blues', cbar_kwargs={'label': ' Freezing Days (2080-2100 minus 1990-2020)'})
plt.title("Change in Freezing Days: 2080-2100 vs 1990-2020")
plt.show()

```
